```python
import requests
import pandas as pd

years = range(1997, 2025)
all_data = []

for year in years:
    url = (
        f"https://api.census.gov/data/timeseries/poverty/saipe"
        f"?get=NAME,SAEMHI_PT"   # state name + median household income
        f"&for=state:*"           # all 50 states + DC
        f"&YEAR={year}"
    )
    response = requests.get(url)
    data = response.json()
    
    # First row is headers, rest is data
    headers = data[0]
    rows = data[1:]
    df = pd.DataFrame(rows, columns=headers)
    df["year"] = year
    all_data.append(df)

# Combine all years
df_all = pd.concat(all_data, ignore_index=True)

# Clean up
df_all = df_all.rename(columns={"NAME": "state_name", "SAEMHI_PT": "median_income"})
df_all["median_income"] = pd.to_numeric(df_all["median_income"])
df_all = df_all[["year", "state_name", "median_income", "state"]]

df_all.to_csv(f"../data/census/saipe_state_median_income.csv", index=False)
print(df_all.head())
```
